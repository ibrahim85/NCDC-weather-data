#<font face='Verdana'>Anomaly Detection</font>#


<font face='Verdana'>
The Global Historical Climatology Network collects weather data from around 80k stations everyday. These measurements provide an incredible source of insight into our changing climate. Using this data, we can determine what the weather is normally like for most places on Earth. We can tell that the average low temperature in New York City on January 11th is 29°F and that the average high temperature in Los Angeles on July 24th is 80°F. Once we know what temperatures to expect on any given day with a certain degree of confidence, we can sift out the uneventful days, leaving only anomalous weather events. <br><br>
Using these criteria, in the second part of our project we track the last 68 years of temperature anomalies for Indian stations. India has about **3805** stations, of which only **85** stations have enough data for analysis. The distribution of these 85 stations is as follows-
</font>

```python
Image(filename='images/Indian stations grouped by state.png',width=2000,height=600) 
```

<font face='Verdana'>
We calculated mean and variance of TMIN and TMAX data for each day of an year for each of these 85 stations using data collected over several years and used that to find anomalies in temperature data of India from 1944-2012. <br>
A day is a cold anomaly for a station if the TMIN on that day is less than (&mu;<sub>TMIN</sub> - k x &sigma;<sub>TMIN</sub>) for that day.<br\>
A day is a warm anomaly for a station if the TMAX on that day is greater than (&mu;<sub>TMAX</sub> + k x &sigma;<sub>TMAX</sub>) for that day.<br\>
k is a parameter using which we control how extreme is our anomaly. 
</font>

#<font face='Verdana'>Partitioning and MDL clustering</font>#


<font face='Verdana'>In the third part of this project, we used Mean Temperature readings available in GSOD Dataset to hierarchially partition the stations into different clusters. We define 'count' as the number of years which have more than 50% readings or TEMP. The splitting is done alternatively between latitudes and longitudes such that sum of counts is same around the split. The stopping criterion used for splitting is either the group has <1000 rows of data or it's width/length is < 5 degrees. 
Here is the initial partioning we get with splitting using KD Trees (~230 clusters)</font>.
<img src='images/before_mdl.png' width=600 height=60/>

<font face='Verdana'>We then used the MDL criterion to merge the stations <br />
If P1 and P2 are neighbors, we check if MDL(P1+P2) < MDL(P1) + MDL(P2). If so, we merge P1 and P2. If not, we check other partitions. We do this till no merging is possible. <br />
We tuned the inequality with an offset to get reasonable number of clusters. Here is the final partioning we get with splitting using KD Trees (~ 14 clusters). </font>
<img src='images/after_mdl.png' width=600 height=60/>

<font face='Verdana'>We see that most of the clusters are formed around equator showing that most of the temperature differences are around it. </font>
